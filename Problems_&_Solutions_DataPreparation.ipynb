#**Problem 13**
# Assing all the missing values using average values 


```python
# Import library 
import pandas as pd

# Create a Url 
url = "https://raw.githubusercontent.com/pmarcelino/datasets/master/titanic.csv"
df = pd.read_csv(url)
df

# Assing all missing values using average values 
df = df.fillna(df.mean())
df
```

#**Problem 14**
# Assingn missing data to specific columns (using the mean)


```python
# Import library 
import pandas as pd 

# Create a Url 
url = "https://raw.githubusercontent.com/pmarcelino/datasets/master/titanic.csv"
df = pd.read_csv(url)
df

#Assing missing data to the column 'Age' using avarege values 
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```
